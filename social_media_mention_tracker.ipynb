```python
posts = [
    "I love learning Python with @code_master!",
    "Is @alice coming to the hackathon?",
    "Shoutout to @code_master for the help today.",
    "Hey @bob, did you see the post by @alice?",
    "Python dictionaries are awesome. @code_master agrees!"
]
```

```python
mentions = {}
```

```python
for i in posts:
    words = i.split()

    for i in words:
          clean_word = i.strip("!?.,")

          if clean_word.startswith("@"):
            if clean_word in mentions:
                mentions[clean_word] += 1
            else:
                mentions[clean_word] = 1
print("--- Mention Count ---")
for user, count in mentions.items():
    print(f"{user}: {count} mentions")
```

```python
Using a dictionary to track multiple instances of an item or object can be very useful when you have ro look for multiple iterations of something
```
